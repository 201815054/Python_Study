```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('./data/ml_source_files/USA_Housing.csv')
```

```python
df.head(3)
```

```python
# # 집값을 예측하는 모델 
# y : 집값(price) - 연속값(continuous data => regression)
# x : income, house age, nums of room, nums of bedroom 
```

```python
y = df['Price']
x = df.drop(['Price','Address'], axis=1) # 집값을 예측해야하니까 집값을 뺌 (예측하는데상관없는것)
```

```python
# train test split (트레인데이터와 테스트데이트 분리)
# train : 모델을 훈련하는데 사용하는 데이터
# test : 모델을 평가하는데 사용하는 데이터 
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
# algorithm 을 갖고 학습시키기 
```

```python
lr = LinearRegression() # 대표적인 회귀모델 
```

```python
# model train
```

```python
lr.fit(x_train,y_train)
```

```python
# predict
```

```python
preds = lr.predict(x_test) #테스트데이터에대해 어떤값을 예측했는지 
preds
```

```python
# evaluation 회귀 모델을 평가할때는
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,preds)
```
